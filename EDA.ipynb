# Exploring Population Trends
# POPULATION GROWTH

```python
# Which countries(top 10) have experienced the most significant population growth over the past decade?
df[["Country","Growth Rate"]].sort_values(by = "Growth Rate", ascending = False).head(10)
plt.figure(figsize=(15, 5))
ax = sns.barplot(x='Country', y = "Growth Rate", data = df[["Country","Growth Rate"]].sort_values(by = "Growth Rate", ascending = False).head(10))
plt.title("Top 10 Most Populated Countries")
for i in ax.containers:
    ax.bar_label(i)
```

```python
# Which countries(top 10) have experienced population decline?
df[["Country","Growth Rate"]].sort_values(by = "Growth Rate").head(10)
plt.figure(figsize=(20, 5))
ax = sns.barplot(x='Country', y = "Growth Rate", data = df[["Country","Growth Rate"]].sort_values(by = "Growth Rate").head(10))
plt.title("Countries Experiencing Most Population Decline")
for i in ax.containers:
    ax.bar_label(i)
```

# POPULATION GROWTH

```python
# How is the population distributed across continents
gb = df.groupby("Continent")[["2022 Population","2020 Population", "2015 Population","2010 Population",
       "2000 Population", "1990 Population", "1980 Population",
       "1970 Population"]].sum().sort_values(by = "Continent")
gb
```

```python
# Visualization of above code
gb = df.groupby("Continent")[["2022 Population"]].sum()
plt.figure(figsize=(20, 5))
ax = sns.barplot(x="Continent", y = "2022 Population", data = gb)
plt.title("Population Distributed Across Continents(2022)")
for i in ax.containers:
    ax.bar_label(i)
```

```python
# Which continents have the highest and lowest population densities?
gb = df.groupby("Continent")
gb[["Density (per km²)"]].max().sort_values(by = "Density (per km²)" )
```

```python
gb = df.groupby("Continent")[["Density (per km²)"]].max().sort_values(by = "Density (per km²)" , ascending = False )
plt.figure(figsize=(15, 4))
ax = sns.barplot(x="Continent", y = "Density (per km²)", data = gb)
plt.title("Continents Having Highest and Lowest Population Densities?")
for i in ax.containers:
    ax.bar_label(i)
```

```python
# Which country has the largest area in km
df[["Country","Area (km²)"]].sort_values(by = "Area (km²)", ascending = False).head(10)
plt.figure(figsize = (20,5))
ax = sns.barplot(x = "Country" , y = "Area (km²)" , data = df[["Country","Area (km²)"]].sort_values(by = "Area (km²)", ascending  = False).head(10))
plt.title("Countries Having The Largest Area in KM")
plt.xticks(rotation = 45, fontsize = 10)
for i in ax.containers:
    ax.bar_label(i)
```
